### 2. 남자를 0, 여자를 1로 가지는 sex_num 컬럼 추가

```python
# 남자 : male, duwk : female
# 남자 :0, 여자 : 1
# 문자타입 범주형 데이터를 숫자타입 범주형 데이터로 변환
data.head()
```

```python
data['sex_num'] = -1
```

```python
# map을 활용한 변환
data['sex_num'] = data["Sex"].map({'male':0, 'female': 1})
```

```python
# 선택한 컬럼의 로우별 값이 male이면 0, female이면 1을 리턴하는 함수
def chage_str_to_num(value):
    if value == 'male':
        return 0
    else :
        return 1
```

### 3. 이름 길이 값을 저장하는 name_length 컬럼 추가

```python
data['name_length'] = 1
```

```python
data['name_length'] = data["Name"].map(len)
```

```python
data
```

```python
# 성별이 영향을 줬는지 여부 확인하기
# 상관계수를 구하면 컬럼간 관계를 유추할 수 있습니다. 성별이 가장많은 영향을 주었음
data[['Survived', 'sex_num']].corr()
```

```python
# 이름길이가 영향을 줬는지 여부 확인하기
# 0.3 이므로 poor or week 약한영향.
data[['Survived', 'name_length']].corr()
```

```python
# 나이는 생존여부에 영향을 미치지 않습니다. 상관계수가 0.07이므로 다른개념으로 봐도 무방
data[['Survived','Age' ]].corr()
```

#### 4. fare 데이터를 소수점 첫 번째 자리에서 반올림한 실수형으로 표기 변경 후 새로운 컬럼으로 추가(fare_ceil)

- 파이썬 내장함수 : round()
- numpy 함수 : np.ceil()

```python
data[['fare_ceil']] = data[['Fare']].round(0)
```

```python
data
```

```python
# np.ceil() : 소수점 첫 자리에 있는 모든 값(사사오입 안함)에 대해서 올림을 적용
# 특정값보다 큰 정수 중 최소값으로 적용
data['Fare'].map(np.ceil)
```

```python
# 어린아이 ~ 중년 직전까지 끊어서보고
# 조건색인으로 0세 ~ 40세
# 이렇게도 가능 data[data["Age"] <=40][['Survived', 'Age']].corr()
df1 = data[data["Age"] <=40]
df1[['Survived', 'Age']].corr()
```
